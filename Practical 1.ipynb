Practical1: Basic numpy & Pands Operations. 

```python
import numpy as np
import pandas as pd
import statistics

# Sample data
data = {'col1': [1, 2, 3, 4, 5],
        'col2': [6, 7, 8, 9, 10],
        'col3': [11, 12, 13, 14, 15]}
df = pd.DataFrame(data)

# NumPy Operations
print("---------- NumPy Operations ----------\n")
col1_arr = df['col1'].to_numpy()
col2_arr = df['col2'].to_numpy()
col3_arr = df['col3'].to_numpy()

print("col1:", col1_arr)
print("col2:", col2_arr)
print("col3:", col3_arr)

print("Sum of col1:", np.sum(col1_arr))
print("Sum of col2:", np.sum(col2_arr))
print("Sum of col3:", np.sum(col3_arr))

print("Mean of col1:", np.mean(col1_arr))
print("Mean of col2:", np.mean(col2_arr))
print("Mean of col3:", np.mean(col3_arr))

print("Median of col1:", np.median(col1_arr))
print("Median of col2:", np.median(col2_arr))
print("Median of col3:", np.median(col3_arr))

print("Standard deviation of col1:", np.std(col1_arr, ddof=1))  # Sample standard deviation
print("Standard deviation of col2:", np.std(col2_arr, ddof=1))
print("Standard deviation of col3:", np.std(col3_arr, ddof=1))

# Pandas Operations
print("\n---------- Pandas Operations ----------\n")

print("Correlation between col1 and col2:", df['col1'].corr(df['col2']))
print("Correlation between col1 and col3:", df['col1'].corr(df['col3']))
print("Correlation between col2 and col3:", df['col2'].corr(df['col3']))

# Statistics Library Operations
print("\n---------- Statistics Library Operations ----------\n")

print("Mean of col1:", statistics.mean(col1_arr.tolist()))
print("Mean of col2:", statistics.mean(col2_arr.tolist()))
print("Mean of col3:", statistics.mean(col3_arr.tolist()))

print("Median of col1:", statistics.median(col1_arr.tolist()))
print("Median of col2:", statistics.median(col2_arr.tolist()))
print("Median of col3:", statistics.median(col3_arr.tolist()))

print("Standard deviation of col1:", statistics.stdev(col1_arr.tolist()))
print("Standard deviation of col2:", statistics.stdev(col2_arr.tolist()))
print("Standard deviation of col3:", statistics.stdev(col3_arr.tolist()))

print("Variance of col1:", statistics.variance(col1_arr.tolist()))
print("Variance of col2:", statistics.variance(col2_arr.tolist()))
print("Variance of col3:", statistics.variance(col3_arr.tolist()))
```
